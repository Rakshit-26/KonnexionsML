````python
import numpy as np
````

````python
x = np.linspace(0,10,10)
y= np.linspace(0,10,10)
````

````python
print(x,y)
````

````python
plt.plot(x,y,marker ='o')
#https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers 
````

````python
plt.plot(x,y,marker ='^')
plt.title("ML konnexions Heros")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

````python
plt.figure(figsize=(6,4))
plt.plot(x,y,marker ='^')
plt.title("ML konnexions Heros")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

````python
x1 = np.linspace(0,10,5)
y1 = np.linspace(0,10,5)

x2 = np.linspace(5,10,5)
y2 = np.linspace(2,8,5)

plt.figure(figsize=(6,4))
plt.plot(x1,y1,marker ='^',label="line 1")
plt.plot(x2,y2,marker ='.',label="line 2")
plt.title("ML konnexions Heros")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
````

````python
x1 = np.linspace(0,10,5)
y1 = np.linspace(0,10,5)

x2 = np.linspace(5,10,5)
y2 = np.linspace(2,8,5)

plt.figure(figsize=(6,4))
plt.plot(x1,y1,marker ='^',label="ML konn",color ='red')
plt.plot(x2,y2,marker ='.',label="GDSC",color='green')
plt.title("ML konnexions Heros")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
````

````python
%matplotlib inline
````

````python
year=[2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
toyota_sales = [1843669, 1496211, 1396837, 1764802, 1893874, 2004373, 2098545, 2106332, 2106332, 2129177, 2128201]
plt.bar(year, toyota_sales, color = 'red', edgecolor='k', label='toyota', alpha=0.4)
plt.xlabel('year')
plt.ylabel('no of unit sold')
plt.show()
````

````python
# plt.style.use('ggplot')
x = [22, 18, 13, 10]
# y = ['maths', 'physics', 'chemistry', 'english']
# plt.pie(x, labels=labels)
plt.pie(x, labels = ['maths', 'physics', 'chemistry', 'english'],  colors =['b','r', 'g', 'm'])
plt.show()
````

````python
#to specify the location of the legend 
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
z = [2,4,6,8,10,12]
plt.plot(x,y, color = 'r', label = "one")
plt.plot(y, color = 'green', label = 'two')     # by default take x =[0,1,2,3,4]
plt.plot(z,linestyle='--', label = 'three')
plt.legend(loc='upper right')
# plt.legend(loc='best')
# plt.legend(['onee', 'two', 'three'])

plt.show()
````
